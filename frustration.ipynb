```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import f1_score, accuracy_score, classification_report

from sklearn.ensemble import RandomForestClassifier
```

```python
df_frustrated = pd.read_excel('./data/фрустрация.xlsx')
df_frustrated.head()
```

```python
df_normal = pd.read_excel('./data/спокойные.xlsx')
df_normal.head()
```

```python
df_frustrated['target'] = 1
df_normal['target'] = 0

df = pd.concat([df_frustrated, df_normal], ignore_index=True)
```

```python
df.head()
```

-------

```python
def cross_validation(model, X, y, n_folds=5, random_state=42, shuffle=True):
    
    X = np.array(X)
    y = np.array(y)
    
    acc_scores = []
    f1_scores = []
    
    skf = StratifiedKFold(n_splits=n_folds, random_state=random_state, shuffle=shuffle)
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        acc_scores.append(accuracy_score(y_test, y_pred))
        f1_scores.append(f1_score(y_test, y_pred))
    
    print('accuracy \t {:.2f} +- {:.2f}'.format(np.mean(acc_scores), np.std(acc_scores)))
    print('f1 score \t {:.2f} +- {:.2f}'.format(np.mean(f1_scores), np.std(f1_scores)))
```

## Все признаки

```python
y = df['target']
X = df.drop(['target', 'Unnamed: 0'], axis=1)
cols = list(df.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, 
                                                    random_state=42, stratify=y)
```

### SVM

```python
svm_model = SVC()

params = {
            'kernel': ['linear'],
            'C': np.arange(0.01, 10, 0.5),
            'probability': [True, False]
         }

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


grid = GridSearchCV(svm_model, 
                    param_grid = params,  
                    n_jobs = -1, 
                    cv = skf.split(X_train, y_train),
                    scoring='accuracy')
```

```python
# linear, c=0.1, probability=False -> acc 0.79
```

```python
svm = SVC(kernel='linear',
          C=0.1,
          probability=False)

svm.fit(X_train, y_train)

y_pred = svm.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

### Random Forest

```python
rfc = RandomForestClassifier(n_estimators=100,
                             criterion='gini',
                             min_samples_split=10,
                             random_state=42)
```

```python
cross_validation(rfc, X, y, n_folds=3)
```
